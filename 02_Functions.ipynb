# Functions

## What is a Function?
- A function is a block of reusable code that performs a specific task.
- Functions help break a program into smaller, manageable parts.
- Instead of writing the same logic again and again, we define a function once and reuse it whenever needed.

## Function Rules

- Function name must start with only alphabets.

In Python, functions are defined using the `def` keyword.

## Syntax
$$
\begin{aligned}
&\text{def } function\_name(parameters): \\
&\qquad statements \\
&\qquad \text{return } value
\end{aligned}
$$

## Types of Functions in Python

Functions in Python are mainly classified into two types:
1. Predefined (Built-in) Functions
2. User-Defined Functions

### 1. Predefined (Built-in) Functions
- Predefined functions are functions that are already available in Python.
- These functions are provided by Python and can be used directly without defining them.

```python
x = 10
name = "Python"

print(x)
print(len(name))
print(type(x))
```

### 2. User-Defined Functions

User-defined functions are functions created by the user to perform a specific task according to the program requirement.

User-defined functions are further classified into different sub-types.
1. Function without parameters
2. Functions with parameters
3. Functions with return

**1. Function without Parameters**

A function without parameters does not take any input values.

```python
def greet():
    print("Hello, Welcome to Python")

greet()
```

**2. Function with Parameters**

A function with parameters accepts input values and uses them to perform operations.

```python
def add(a, b):
    print("Sum =", a + b)

add(10, 20)
```

**3. Function with Return Value**

A function with return value sends the result back to the caller using the return statement.

```python
def square(n):
    return n * n

result = square(5)
print("Square =", result)
```

## Positional Arguments

Positional arguments are used to store the values in the same sequence as defined in the function.

```python
def add(a, b):
    print(a, b)

add(10, 20)
```

## Keyword Arguments

Keyword arguments allow passing values using parameter names.

```python
def add(a, b):
    print(a, b)

add(b=10, a=20)
```

## Default Arguments

- Default arguments are used to assign a default value to a parameter.
- If no value is passed, the default value is used.

```python
def name(n="user"):
    print("Hi", n)

name()
```

## Arbitrary Arguments (*)

- Arbitrary arguments are used when the number of arguments is not fixed.
- A single variable can take more than one value.

```python
def add(*numbers):
    print(sum(numbers))

add(10, 20, 30)
```

## Local and Global Variable

```python
n = 5
def sample():
    n = 10
    print(n)

print(n)
sample()
```
