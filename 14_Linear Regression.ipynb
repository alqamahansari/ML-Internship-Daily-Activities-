## Regression

Regression is used to model the relationship between **input variables
(features)** and a **continuous output variable**, with the goal of
predicting numerical outcomes.

## Linear Regression

### What is Linear Regression?

- Linear regression is a **supervised machine learning algorithm** used to predict **continuous numerical values**.

- It models the relationship between a **dependent variable (Y)**, and one or more **independent variables (X)** using a **straight line**.

### Examples

- Predicting marks based on study hours  
- Predicting house price based on area  


## Types of Linear Regression

1. **Simple Linear Regression** 
2. **Multiple Linear Regression**  

## Simple Linear Regression

Simple Linear Regression uses **only one independent variable (X)** to
predict a **dependent variable (Y)**.

The goal is to find a **straight line** that best fits the data points.

### Equation

$$
\begin{aligned}
y &= b_0 + b_1 x
\end{aligned}
$$

where,
$$
\begin{aligned}
b_0 &\rightarrow \text{Intercept} \\
b_1 &\rightarrow \text{Slope}
\end{aligned}
$$

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    "Study_Hours":[1,2,3,4,5],
    "Marks":[2,4,5,4,5]
})
df
```

```python
X=df[["Study_Hours"]]
y=df["Marks"]
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
X_line = pd.DataFrame(
    {'Study_Hours' : np.linspace(X.min()[0], X.max()[0], 100)}
)
y_line=model.predict(X_line)
```

```python
X_line
```

```python
print("Slope (m)", model.coef_[0])
print("Intercept (b):",model.intercept_)
```

```python
X_new=pd.DataFrame({'Study_Hours':[4.0]})
y_new=model.predict(X_new)
print("Predict Value: ",y_new)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(X_line, y_line)
plt.scatter(X_new,y_new)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression: Best-Fit Line")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X)
mse = mean_squared_error(y,y_pred)
```

```python
print("MSE:", mse)
```

## Multiple Linear Regression

Multiple Linear Regression uses **more than one independent variable**
(\( x_1, x_2, x_3, \dots, x_n \)) to predict the dependent variable \( y \).

The goal is to find a **hyperplane** that best fits the data.

### Equation

$$
\begin{aligned}
y &= b_0 + b_1 x_1 + b_2 x_2 + \cdots + b_n x_n
\end{aligned}
$$

Where:
$$
\begin{aligned}
b_0 &\rightarrow \text{Intercept} \\
b_1, b_2, \dots, b_n &\rightarrow \text{Coefficients of independent variables}
\end{aligned}
$$


### Simple Linear Regression (SLR) vs Multiple Linear Regression (MLR)

- **SLR** fits a **straight line** in 2D space (one input variable)
- **MLR** fits a **plane / hyperplane** in higher-dimensional space
  (multiple input variables)


### Example (Simple Linear Regression)

### Dataset

| Study Hours (X) | Marks (Y) |
|-----------------|-----------|
| 1               | 2         |
| 2               | 4         |
| 3               | 5         |
| 4               | 4         |
| 5               | 5         |


### Best Fit Line

The equation of the best-fit line is:

$$
\begin{aligned}
y &= mx + c
\end{aligned}
$$

### Formula for Slope and Intercept

#### Slope \( m \)

$$
\begin{aligned}
m
&= \frac{n\sum xy - (\sum x)(\sum y)}
         {n\sum x^2 - (\sum x)^2}
\end{aligned}
$$

### Intercept \( c \)

$$
\begin{aligned}
c
&= \frac{\sum y - m\sum x}{n}
\end{aligned}
$$


## Computation Table

| X | Y | XY | \(X^2\) |
|---|---|----|---------|
| 1 | 2 | 2  | 1       |
| 2 | 4 | 8  | 4       |
| 3 | 5 | 15 | 9       |
| 4 | 4 | 16 | 16      |
| 5 | 5 | 25 | 25      |
| **Σ** | **–** | **66** | **55** |

$$
\begin{aligned}
\sum x &= 15, \\
\sum y &= 20, \\
\sum xy &= 66, \\
\sum x^2 &= 55
\end{aligned}
$$

## Calculating the Slope

$$
\begin{aligned}
m
&= \frac{5(66) - (15)(20)}
         {5(55) - (15)^2} \\
&= \frac{330 - 300}{275 - 225} \\
&= \frac{30}{50} \\
&= 0.6
\end{aligned}
$$

## Calculating the Intercept
$$
\begin{aligned}
c
&= \frac{\sum y - m \sum x}{n} \\
&= \frac{20 - 0.6(15)}{5} \\
&= \frac{20 - 9}{5} \\
&= \frac{11}{5} \\
&= 2.2
\end{aligned}
$$

## Final Best-Fit Line
$$
\begin{aligned}
y &= mx + c \\
y &= 0.6x + 2.2
\end{aligned}
$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv(r"C:\Resource\Sample data\MLR_Study_marks.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[["Study_Hours", "Attendance", "Sleep_Hours"]]
y = df["Marks"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 40)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

print("\nModel Intercept:", model.intercept_)
print("Model Coefficients:", model.coef_)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Marks")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nMean Squared Error:", mse)
print("R^2 Score:", r2)
```
