# Covariance
covariance tells you whether two variables move together.
- positive covariance --> X variable increases y also increases
- Negative covariance --> X variable increases y also decreases
- Zero covariance --> no relation b/w the variables
- NOTE:: covariance shows only direction not strength

```python
import pandas as pd
df=pd.read_csv(r"C:\Resource\Sample data\Study.csv")
df
```

```python
df.cov() # find a covariance , if the diagonal has positive values, range is (-infinity ,+ infinity)
```

# Normalization
Normalization is a technique used to scale the values of features into a small range, usually 0 to 1. It makes all features equal in scale to the model does not get confused by big and small numbers.

### use
it have different ranges big numbers dominate small numbers models work better when all values are in same range training becomes faster and more accurate

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pandas as pd
```

```python
s=pd.Series([1000,2000,3000,4000,9000])
s
```

```python
s.min()
```

```python
s.max()
```

```python
sn=(s-s.min())/(s.max()-s.min())
sn
```

```python
sn=MinMaxScaler().fit_transform(s.values.reshape(-1,1))
print(sn)
```

```python
data= {
    "temp": [30,32,34,35,36],   
    "humidity":[40,39,34,28,27]
}
df=pd.DataFrame(data)
df
```

```python
norm_data=MinMaxScaler().fit_transform(df) # to normalize the code
n_df= pd.DataFrame(norm_data,columns=df.columns) # it gives the data in an array
norm_data
```

```python
n_df # it displays in a structured format
```

# Standardization
Standardization or Z score normalization is a scaler technique that transforms data so that : Mean = 0,standard deviation= 1. It shows how far each value is from the mean.

### Use
makes data centered around zero
works well when data follows a normal distribution
less affected by outliers compared to normalization
Helps the model learn more smoothly
## Algorithms
lOgistic Regression pCA sVM linear regression neural Networks

range is -->(-infinity,+infinity)
z= (x=mue)/sigma, mue=mean, sigma= standard deviation

```python
import pandas as pd 
s=pd.Series([1000,2000,3000,4000,9000])
s
```

```python
s.mean()
```

```python
s.std()
```

```python
z_score=(s-s.mean())/s.std()
print(z_score)
```

```python
z_score=(s-s.mean())/s.std(ddof=0)
print(z_score)
```

```python
sn=StandardScaler().fit_transform(s.values.reshape(-1,1)) # standard deviation (n population data)
print(sn)
```

```python
data= {
    "temp": [30,32,34,35,36],   
    "humidity":[40,39,34,28,27]
}
df=pd.DataFrame(data)
df
```

```python
norm_data=StandardScaler().fit_transform(df) # to standardize the code
n_df= pd.DataFrame(norm_data,columns=df.columns) # it gives the data in an array
norm_data
```
