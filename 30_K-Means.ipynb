# K-Means

## What is Clustering?

- **Clustering** is an **unsupervised learning technique** that groups similar data points into clusters **without using labeled data**.
- Data points within the same cluster are **more similar to each other** than to points in other clusters.


## What is K-Means Clustering?

- **K-Means Clustering** is an **unsupervised learning algorithm** that divides a dataset into **K distinct clusters**.
- It works by **minimizing the distance** between data points and their respective **cluster centroids**.
- Each cluster is represented by a **centroid**, which is the **mean of all data points** in that cluster.


## Why is it called K-Means?

- **K** → Number of clusters  
- **Means** → The **average (centroid)** of points in each cluster



## How K-Means Works (Algorithm Steps)

1. Choose the number of clusters **K**.
2. Initialize **K centroids randomly**.
3. Calculate the distance between each data point and the centroids.
4. Assign each data point to the **nearest centroid**.
5. Update the centroids by calculating the **mean** of all points in each cluster.
6. Repeat steps **3–5** until the centroids **do not change**.


## Formulas Used in K-Means

### Euclidean Distance
$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

### Centroid Formula
$$
\text{Centroid} = \left( \frac{\sum x}{n}, \frac{\sum y}{n} \right)
$$


## Advantages of K-Means

- Simple and easy to understand  
- Fast and efficient  
- Works well for large datasets  


## Limitations of K-Means

- Value of **K** must be chosen manually  
- Sensitive to outliers  
- Different initial centroids may give different results  

## Example: K-Means Clustering (Step-by-Step)

### Step 1: Dataset

| Point | x | y |
|------|---|---|
| A | 1 | 1 |
| B | 1 | 2 |
| C | 4 | 4 |
| D | 5 | 4 |



Number of clusters:  
$$
K = 2
$$

### Step 2: Initialize Centroids (Randomly)

Let the initial centroids be:

- **Centroid 1 (C₁)** = A(1, 1)  
- **Centroid 2 (C₂)** = C(4, 4)


### Step 3: Calculate Distance (Euclidean Distance)

#### Distance from Point A (1,1)

$$
d(A, C_1) = \sqrt{(1-1)^2 + (1-1)^2} = 0
$$

$$
d(A, C_2) = \sqrt{(4-1)^2 + (4-1)^2} = \sqrt{18} \approx 4.24
$$

➡ A is assigned to **Cluster 1**


#### Distance from Point B (1,2)

$$
d(B, C_1) = \sqrt{(1-1)^2 + (2-1)^2} = \sqrt{1} = 1
$$

$$
d(B, C_2) = \sqrt{(4-1)^2 + (4-2)^2} = \sqrt{13} \approx 3.6
$$

➡ B is assigned to **Cluster 1**


#### Distance from Point C (4,4)

$$
d(C, C_1) = \sqrt{(1-4)^2 + (1-4)^2} = \sqrt{18} \approx 4.24
$$

$$
d(C, C_2) = \sqrt{(4-4)^2 + (4-4)^2} = 0
$$

➡ C is assigned to **Cluster 2**


#### Distance from Point D (5,4)

$$
d(D, C_1) = \sqrt{(1-5)^2 + (1-4)^2} = \sqrt{25} = 5
$$

$$
d(D, C_2) = \sqrt{(4-5)^2 + (4-4)^2} = \sqrt{1} = 1
$$

➡ D is assigned to **Cluster 2**

## Cluster Assignment Table (Iteration 1)

| Point | Distance to C₁ | Distance to C₂ | Assigned Cluster |
|------|----------------|----------------|------------------|
| A (1,1) | 0 | 4.24 | C₁ |
| B (1,2) | 1 | 3.6 | C₁ |
| C (4,4) | 4.24 | 0 | C₂ |
| D (5,4) | 5 | 1 | C₂ |


## Step 4: Update Centroids

### New Centroid for Cluster 1 (A, B)

$$
C_1 = \left( \frac{1+1}{2}, \frac{1+2}{2} \right) = (1, 1.5)
$$

### New Centroid for Cluster 2 (C, D)

$$
C_2 = \left( \frac{4+5}{2}, \frac{4+4}{2} \right) = (4.5, 4)
$$


## Step 5: Repeat Distance Calculation (Iteration 2)

### Distance from Point A (1,1)

$$
d(A, C_1) = \sqrt{(1-1)^2 + (1-1.5)^2} = 0.5
$$

$$
d(A, C_2) = \sqrt{(4.5-1)^2 + (4-1)^2} = \sqrt{12.25 + 9} \approx 4.6
$$

➡ A remains in **Cluster 1**


### Distance from Point B (1,2)

$$
d(B, C_1) = \sqrt{(1-1)^2 + (1.5-2)^2} = 0.5
$$

$$
d(B, C_2) = \sqrt{(4.5-1)^2 + (4-2)^2} = \sqrt{12.25 + 4} \approx 4.03
$$

➡ B remains in **Cluster 1**


### Distance from Point C (4,4)

$$
d(C, C_1) = \sqrt{(1-4)^2 + (1.5-4)^2} = \sqrt{9 + 6.25} \approx 3.9
$$

$$
d(C, C_2) = \sqrt{(4.5-4)^2 + (4-4)^2} = 0.5
$$

➡ C remains in **Cluster 2**

## Distance from Point D (5,4) — Iteration 2

$$
d(D, C_1) = \sqrt{(1-5)^2 + (1.5-4)^2}
           = \sqrt{16 + 6.25}
           = \sqrt{22.25}
           \approx 4.71
$$

$$
d(D, C_2) = \sqrt{(4.5-5)^2 + (4-4)^2}
           = \sqrt{0.25}
           = 0.5
$$

➡ D remains in **Cluster 2**


## Distance Table (Iteration 2)

| Point | Distance to C₁ | Distance to C₂ | New Cluster |
|------|----------------|----------------|-------------|
| A (1,1) | 0.5 | 4.6 | C₁ |
| B (1,2) | 0.5 | 4.03 | C₁ |
| C (4,4) | 3.9 | 0.5 | C₂ |
| D (5,4) | 4.71 | 0.5 | C₂ |


## Step 6: Final Clusters

- **Cluster 1**  
  $$
  A(1,1),\; B(1,2)
  $$

- **Cluster 2**  
  $$
  C(4,4),\; D(5,4)
  $$




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X= np.array([
    [1, 1], #A
    [1,2], #B
    [4, 4], #C
    [5,4] #D
])
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Cluster labels:", labels)
print("Centroids: \n", centroids)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("K-Means Clustering Example")
plt.show()
```

# Why Do We Need the Elbow Method?

- K-Means requires the value of **K** beforehand.
- Choosing the wrong **K** leads to poor clustering.
- The **Elbow Method** helps us find the optimal **K** value.


## Elbow Method

The **Elbow Method** is a technique used to determine the optimal number of clusters in K-Means can be found by plotting the **Within-Cluster Sum of Squares (WCSS)** against different values of **K** and identifying the point where the decrease becomes slower — this is known as the **Elbow Point**.


## WCSS (Within-Cluster Sum of Squares)

WCSS measures how close the data points are to their respective cluster centroid.

$$
WCSS = \sum ||x - \mu||^2
$$

Where:
- \( x \) = data point  
- \( mu \) = cluster centroid  

**Lower WCSS → Better Clustering** <br>
**WCSS decreases as K increases**


# Steps in Elbow Method

1. Apply K-Means for different values of **K**
2. Calculate WCSS for each **K**
3. Plot **K vs WCSS**
4. Identify the elbow point
5. Select that **K** as the optimal number of clusters


# Elbow Method – Graph Explanation

- Sharp decrease in WCSS → Good clustering
- Small decrease after a point → Minor improvement
- That bending point is the **Elbow**

# Elbow Method – Step-by-Step Example


## Step 1: Dataset

| Point | X | Y |
|-------|---|---|
| A     | 1 | 1 |
| B     | 1 | 2 |
| C     | 4 | 4 |
| D     | 5 | 4 |


## Important Formulas

Distance Formula:

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

Centroid Formula:

$$
c = \left(\frac{\sum x}{n}, \frac{\sum y}{n}\right)
$$

Within-Cluster Sum of Squares (WCSS):

$$
WCSS = \sum \|x_i - \mu\|^2
$$


## Step 2: Apply K-Means for Different Values of K


# Case 1: K = 1

### Centroid Calculation

Since K = 1, all points belong to one cluster.

$$
c = \left(\frac{1 + 1 + 4 + 5}{4}, \frac{1 + 2 + 4 + 4}{4}\right)
$$

$$
c = \left(\frac{11}{4}, \frac{11}{4}\right)
$$

$$
c = (2.75, 2.75)
$$


### Distance Calculation (Squared Distances)

| Point | Calculation | Value |
|-------|-------------|-------|
| A | $(1 - 2.75)^2 + (1 - 2.75)^2$ | 6.125 |
| B | $(1 - 2.75)^2 + (2 - 2.75)^2$ | 3.625 |
| C | $(4 - 2.75)^2 + (4 - 2.75)^2$ | 3.125 |
| D | $(5 - 2.75)^2 + (4 - 2.75)^2$ | 6.625 |


### WCSS for K = 1

$$
WCSS = 6.125 + 3.625 + 3.125 + 6.625
$$

$$
WCSS_{K=1} = 19.5
$$


# Case 2: K = 2

Cluster 1: {A, B}  
Cluster 2: {C, D}

## Centroid Calculation

For Cluster 1:

$$
\mu_1 = \left(\frac{1+1}{2}, \frac{1+2}{2}\right)
$$

$$
\mu_1 = (1, 1.5)
$$

For Cluster 2:

$$
\mu_2 = \left(\frac{4+5}{2}, \frac{4+4}{2}\right)
$$

$$
\mu_2 = (4.5, 4)
$$


## Distance Calculation

### Cluster 1

| Point | Calculation | Value |
|--------|-------------|--------|
| A | $(1-1)^2 + (1-1.5)^2$ | 0.25 |
| B | $(1-1)^2 + (2-1.5)^2$ | 0.25 |

$$
WCSS_{Cluster1} = 0.25 + 0.25 = 0.5
$$


### Cluster 2

| Point | Calculation | Value |
|--------|-------------|--------|
| C | $(4-4.5)^2 + (4-4)^2$ | 0.25 |
| D | $(5-4.5)^2 + (4-4)^2$ | 0.25 |

$$
WCSS_{Cluster2} = 0.25 + 0.25 = 0.5
$$


## Total WCSS for K = 2

$$
WCSS_{K=2} = 0.5 + 0.5 = 1
$$


# Case 3: K = 3

Clusters:
- Cluster 1 → {A}
- Cluster 2 → {B}
- Cluster 3 → {C, D}


## Centroids

- A → (1, 1)
- B → (1, 2)
- (C, D) →

$$
\mu_3 = \left(\frac{4+5}{2}, \frac{4+4}{2}\right)
$$

$$
\mu_3 = (4.5, 4)
$$


## Distance Calculation

| Point | Value |
|--------|--------|
| A | 0 |
| B | 0 |
| C | 0.25 |
| D | 0.25 |

$$
WCSS_{K=3} = 0 + 0 + 0.25 + 0.25
$$

$$
WCSS_{K=3} = 0.5
$$

# Step 3: Compare WCSS Values

| K | WCSS |
|---|------|
| 1 | 19.5 |
| 2 | 1.0 |
| 3 | 0.5 |


# Step 4: Elbow Decision

- Huge drop from **K = 1 → K = 2**
- Very small drop from **K = 2 → K = 3**

This indicates that adding more clusters after **K = 2** does not significantly reduce WCSS.


# Step 5: Final K Value

$$
K = 2
$$


# Final Conclusion

The optimal number of clusters using the **Elbow Method** is:

$$
\boxed{K = 2}
$$

```python
wcss = []

for k in range(1, 4):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
for i, val in enumerate(wcss, start=1):
    print (f"K = {i} , WCSS = (val)")
```

```python
plt.plot(range(1, 4), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
drops=[]

for i in range(len(wcss) - 1):
    drops.append(wcss[i]- wcss[i+1])

#Step 3: Fix K (maximum drop elbow)
optimal_k = drops.index(max(drops)) + 2
```

```python
print("WCSS values:", wcss)
print("WCSS drops:", drops)
print("Optimal K value:", optimal_k)
```
