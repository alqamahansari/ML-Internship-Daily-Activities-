# File Handling

```python
f= open('abc.txt','w')
print("File name:",f.name)
print("File Mode:",f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

```python
with open("abc.txt", "r") as file:
    content= file.read()
print(content)
```

```python
with open("abc.txt", "r") as file:
    content= file.read(3)
print(content)
```

```python
with open("abc.txt", "r") as file:#used to close the file automatically after use
    content= file.readlines()#read complete data into list 
print(content)
```

```python
with open("abc.txt", "r") as file:#used to close the file automatically after use
    content= file.readline()#read the first line of data
print(content)
```

```python
with open("abc.txt", "w") as file:#used to close the file automatically after use
    content= file.write("Alqamah")# write the string data in the file
```

```python
l=["Mohammad\n","Alquamah\n","Ansari\n"]# write the list data in a file
with open("abc.txt", "w") as file:#used to close the file automatically after use 
    content= file.writelines(l)
```

```python
with open("abc.txt", "a") as file:#used to close the file automatically after use
    content= file.write("nalllllllllllllllllllllllll")# add the data without removing the previous content
```

```python
import csv # import csv module
with open("ansari.csv", "w",newline="") as file:
    s= csv.writer(file)
    s.writerow(["Name","Age","City"])
    s.writerow(["Ansari",25,"vns"])
```

```python
import csv # import csv module
with open("ansari.csv", "r") as file:# using reader function rows
    s= csv.reader(file)
    for row in s:
        print(row)
```
